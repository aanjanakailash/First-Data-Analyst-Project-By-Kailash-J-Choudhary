```python
import pandas as pd
kp=pd.read_csv("C:/Users/Kailash J Choudhary/Desktop/powerbi_industry_sales_loss_analysis_project/messy_sales_orders_5060_rows_csv.csv")
product=pd.read_csv("C:/Users/Kailash J Choudhary/Desktop/powerbi_industry_sales_loss_analysis_project/products_lookup_csv.csv")

```

```python
print(kp.isnull().sum())
```

```python
print(kp.duplicated().sum())
```

```python
#count and remove duplicates values 
print("Before:", kp.duplicated().sum())
kp = kp.drop_duplicates()
print("After:", kp.duplicated().sum())
print(kp.shape)
```

```python
#remove extra space and set city title
kp["City"]=kp["City"].str.title().str.strip()
print(kp.head())
```

```python
print(kp["Order_Date"].head(10))
print(kp["Order_Date"].isnull().sum())
kp["Order_Date"]=pd.to_datetime(
    kp["Order_Date"],
    errors="coerce",
    dayfirst=True
)
```

```python
1. Remove duplicate rows
2. Fix text format
3. Fix dates
4. Replace Discount null with 0
5. Fill Unit_Cost null from products_lookup
6. Handle Customer_Rating null
7. Remove/flag negative Quantity
8. Fix negative Marketing_Cost
9. Check returned orders without reason
```

```python
print(kp["Discount_Percent"].head())
kp.loc[
    kp["Discount_Percent"].isnull(),
    ["Order_ID", "Discount_Percent"]
]
kp.loc[
        kp["City"]=="jaipur",
        ["Order_ID","City"]]
```

```python
kp.loc[
        kp["City"]=="Jaipur",
        ["Order_ID","City"]]
```

```python
print(kp["Discount_Percent"].isnull().sum())
kp.head()
kp["Discount_Percent"]=kp["Discount_Percent"].fillna(0)
```

```python
kp["Customer_Rating"].isnull().sum()
avg_rating=kp["Customer_Rating"].mean()
kp["Customer_Rating"]=kp["Customer_Rating"].fillna(avg_rating)
kp["Customer_Rating"].isnull().sum()
kp.loc[
        kp["Customer_Rating"]<0
        ]
    
```

```python
# convert all  negative Quantity in to positive quantity
kp.loc[
    kp["Quantity"] < 0,
    ["Order_ID", "Product_Name", "Quantity", "Order_Status"]
]
kp["Quantity"] = kp["Quantity"] .abs()
kp.loc[
    kp["Quantity"] < 0,
    ["Order_ID", "Product_Name", "Quantity", "Order_Status"]
]
```

```python
kp=kp.merge(
        product[["Product_ID","Base_Unit_Price"]],
        on="Product_ID",
        how="left"
)
```

```python
kp["Unit_Cost"]=kp["Unit_Cost"].fillna(kp["Base_Unit_Price"])
kp=kp.drop(columns=["Base_Unit_Price"])
```

```python
print(kp["Unit_Cost"].isnull().sum())
```

```python
#Fix negative Marketing_Cost
kp.loc[
    (kp["Marketing_Cost"]<0).sum(),
      ["Marketing_Cost"]
    ]
kp["Marketing_Cost"]=kp["Marketing_Cost"].abs()
```

```python
kp.loc[
    kp["Marketing_Cost"]<0,
      ["Marketing_Cost"]
    ]
```
